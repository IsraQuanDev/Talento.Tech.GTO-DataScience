```python
import matplotlib.pyplot as plt


conjunto1 = [0, 2, 4, 8, 16, 32, 64]

conjunto2 = [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0]

plt.plot(conjunto1, conjunto2, label='Caso 1', linestyle='dotted', color='green')

plt.plot([0,64], [0,3], label='Caso 2', linestyle='dashed', color='orange')

plt.legend()

plt.title('Gráfica de Muestra')

plt.xlabel('Nombre del eje X')

plt.ylabel('Nombre del eje Y')

plt.grid(True)

plt.show()
```

# Gráfico de Dispersión

```python
import matplotlib.pyplot as plt
import numpy as np

N = 100

#Generamos 100 coordenadas de X
x1 = np.random.rand(N)
#Generamos 100 coordenadas de Y
y1 = np.random.rand(N)
#Generamos 100 coordenadas de X
x2 = np.random.rand(N)
#Generamos 100 coordenadas de Y
y2 = np.random.rand(N)
#Graficamos
plt.scatter(x1, y1, color='yellow', alpha=0.7, label='Puntos Amarillos')
plt.scatter(x2, y2, color='purple', alpha=0.7, label='Puntos Morados')
plt.legend(loc='best')
plt.show()
```

# Gráfica de Pastel

# Gráfico de Dispersión para visualizar la relación entre el total de la cuenta y la propina.

```python
#Gráfico de Dispersión
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(8,6))

sns.lmplot(data=tips, x='total_bill', y='tip', hue='time', legend=False)
plt.title('Relación entre el total de la cuenta y la propina', fontsize=14)
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.legend(title='Momento del día')

plt.show()
```

# Gráfico de Histograma: Mostrar las distribuciones de los totales de las cuentas



```python
plt.figure(figsize=(8,6))
sns.histplot(tips['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Distribución total de la cuenta')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Frecuencia')
```
