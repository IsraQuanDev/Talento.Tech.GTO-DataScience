```python

dstitanic = sns.load_dataset('titanic')
dstitanic
```

```python
#Función "Fillna" rellena los valores nulos con un valor predeterminado

dstitanic['deck'].fillna('C', inplace=True)
dstitanic
```

```python
#Función bfill: rellena NaN con los valores de la siguiente fila

dstitanic['age'].fillna(method="bfill", inplace=True)
dstitanic

```
